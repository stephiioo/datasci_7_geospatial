```python
# load in packages

import geopandas as gpd
from shapely.geometry import Point
import geoplot
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

## Dataset 1: Cancer rates

```python
cancer = gpd.read_file('/content/Cancer_Rates.geojson')
cancer
```

```python
# check for missing values

missing_values = cancer.isna().sum()

# calculate the percentage of missing values
percentage_missing = (missing_values / len(cancer)) * 100

# Display of the results
print("Missing Values:")
print(missing_values)
print("\nPercent of Missing Values:")
print(percentage_missing)
```

```python
# Remove duplicate rows

cancer_no_duplicates = cancer.drop_duplicates()

# Check for duplicate columns

duplicate_columns = cancer_no_duplicates.columns[cancer_no_duplicates.columns.duplicated()]

# Remove duplicate columns

cancer_no_duplicate_columns = cancer_no_duplicates.loc[:, ~cancer_no_duplicates.columns.duplicated()]

# Print the result or perform further operations

print("Duplicate columns:", duplicate_columns)
print("DataFrame without duplicate columns shape:", cancer_no_duplicate_columns.shape)
```

```python
# Clean the column names for the DataFrame named 'cancer'

cancer.columns = cancer.columns.str.lower()
cancer.columns = cancer.columns.str.strip()
cancer.columns = cancer.columns.str.replace(' ', '_')
cancer.columns = cancer.columns.str.replace('[^a-zA-Z0-9_]', '', regex=True)

# Print the updated column names

print(cancer.columns)
```

```python
cancer.plot("prostate_c", legend=True)
```
